```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy import stats
import matplotlib.pyplot as plt

# Membaca data dari file CSV
data = pd.read_csv('DataCoSupplyChainDataset.csv', encoding="latin1", sep=";")

# Persiapan data untuk analisis regresi
X = data[['Order Item Discount', 'Order Item Quantity', 'Product Price']]
y = data['Order Item Total']

# Membuat model regresi linear
model = LinearRegression()
model.fit(X, y)

# koefisien regresi
coefficients = pd.DataFrame({'Variable': X.columns, 'Coefficient': model.coef_})
print("Pengaruh Variabel Independen (dalam angka):")
print(coefficients)
print()

# R-squared
r_squared = model.score(X, y)
print("R-squared:", r_squared)
print()

# 5. Uji Korelasi Antar Variabel
correlation_matrix = data[['Order Item Discount', 'Order Item Quantity', 'Product Price', 'Order Item Total']].corr()
print("Matriks Korelasi:")
print(correlation_matrix)
print()

plt.figure(figsize=(20, 6))
plt.subplot(1, 3, 1)
plt.scatter(data['Order Item Discount'], y)
plt.xlabel('Order Item Discount')
plt.ylabel('Order Item Total')
plt.title('Scatter Plot: Order Item Discount vs Order Item Total')

plt.subplot(1, 3, 2)
plt.scatter(data['Order Item Quantity'], y)
plt.xlabel('Order Item Quantity')
plt.ylabel('Order Item Total')
plt.title('Scatter Plot: Order Item Quantity vs Order Item Total')

plt.subplot(1, 3, 3)
plt.scatter(data['Product Price'], y)
plt.xlabel('Product Price')
plt.ylabel('Order Item Total')
plt.title('Scatter Plot: Product Price vs Order Item Total')

plt.tight_layout()
plt.show()
```
